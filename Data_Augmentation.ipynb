```python
from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img('/content/cat jpg.jpg', target_size=(200,200))
```

```python
img
```

```python
import matplotlib.pyplot as plt

plt.imshow(img)
```

```python
datagen = ImageDataGenerator(
    rotation_range=0.2, # define rotation
    shear_range=0.2, # image distortion
    zoom_range=0.2, # zoom in or zoom out
    horizontal_flip=True, # flipping image horizontally
    width_shift_range=0.2, #
    height_shift_range=0.2,
)
```

```python
img = image.img_to_array(img)
```

```python
img.shape
```

```python
input_batch = img.reshape(1,200,200,3)
```

```python
i = 0

for output in datagen.flow(input_batch, batch_size=1, save_to_dir='aug'): # in aug file there will be 10 images of cat
    i = i + 1
    if i == 10:
        break
```
