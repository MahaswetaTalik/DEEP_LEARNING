```python
import pandas as pd
```

```python
import numpy as np

date_range = pd.date_range(start="2024-01-01", periods=50, freq='D')

np.random.seed(0)
temps = np.random.normal(loc=30, scale=5, size=50)

df = pd.DataFrame({
    'date': date_range,
    'meantemp': temps
})

print(df)
```

```python
df = df[['date','meantemp']]
df
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['date'],df['meantemp'],color='red')
```

```python
x1 = df['meantemp'].ewm(alpha=0.9).mean()
x1
```

```python
df['ewma'] = x1
df
```

```python
plt.scatter(df['date'], df['meantemp'], color='red')
plt.plot(df['date'], x1, color='black', label='alpha = 0.98')
plt.legend()
plt.show()
```
